 Write 5 different prompts for the SAME task:

  - Improved

  - Detailed

  - Creative

  - With clear constraints



## Setup Enviroment

```python
%pip install python-dotenv
%pip install google-genai
```

```python
import os
from dotenv import load_dotenv
from google import genai
from google.genai import types

load_dotenv()
gemini_key = os.getenv("GEMINI_API_KEY")

config = types.GenerateContentConfig(
    temperature= 0.7,
    max_output_tokens= 300,
)

client =genai.Client(api_key=gemini_key)
```

## Basic Prompting

```python
# Basic Prompting
basic_prompt = "write a excuse message" # Only Task

basic_response = client.models.generate_content(
    model='gemini-3.1-flash-lite',
    contents= basic_prompt,
    config=config
)

print(f'AI Response With basic prompt:\n {"-"*30}\n{basic_response.text}\n {"-"*30}')
```

## Improved Prompting

```python
# Improved Prompting

# Here I will increase the specificity of the prompt.
Improved_prompt = "You are a professional employee. Write a short excuse message to your manager." # Putting "Role", "Task", and "To Whom"

Improved_response = client.models.generate_content(
    model='gemini-3.1-flash-lite',
    contents= Improved_prompt,
    config=config
)

print(f'AI Response With improved prompt:\n {"-"*30}\n{Improved_response.text}\n {"-"*30}')
```

## Detailed Prompting

```python
# Detailed Prompting

Detailed_prompt = "You are a professional employee. Write an excuse message to your manager, explaining that you will be late to work because of a family emergency,\
      and you will make up for the lost time by working extra hours this week." # Putting "Role", "Task", and "To Whom","Reason", and "Solution"

Detailed_response = client.models.generate_content(
    model='gemini-3.1-flash-lite',
    contents= Detailed_prompt,
    config=config
)

print(f'AI Response With detailed prompt:\n {"-"*30}\n{Detailed_response.text}\n {"-"*30}')
```

## Creative Prompt

```python
# Creative Prompting

# Putting "Role", "Task", and "To Whom","Reason", "Solution", and "Tone" in the prompt to make it more creative and empathetic. 
Creative_prompt = "You are a professional employee. Write an excuse message to your manager, explaining that you will be late to work because of a family emergency,\
      and you will make up for the lost time by working extra hours this week. You need to be more empathetic and talking like you harry up with some mistaypes that represnt your anixity."

Creative_response = client.models.generate_content(
    model='gemini-3.1-flash-lite',
    contents= Creative_prompt,
    config=config
)

print(f'AI Response With creative prompt:\n {"-"*30}\n{Creative_response.text}\n {"-"*30}')
```

## Prompting with clear constraints

```python
# Prompting with clear constraints

const_prompt = "You are a professional employee. Write an excuse message to your manager, explaining that you will be late to work because of a family emergency,\
      and you will make up for the lost time by working extra hours this week. Addionly, you need to make it only with 50 words, and it should be polite and professional witout be anxious."

# Putting "Role", "Task", and "To Whom","Reason", "Solution", and "Constraints" 

const_response = client.models.generate_content(
    model='gemini-3.1-flash-lite',
    contents= const_prompt,
    config=config
)

print(f'AI response with clear constraints prompt:\n {"-"*30}\n{const_response.text}\n {"-"*30}')
```

## **Comparison & Conclusion**

| Prompt | Result |
|---|---|
| Basic | Didn't give a usable message, the model just asked me questions instead |
| Improved | Usable but too generic, gave a few unrelated template options |
| Detailed | Specific and professional, had the right context |
| Creative | Good emotional tone, but it didn't actually include the "anxious typos" I asked for |
| Constrained | The most polished one, hit the tone I wanted, came in at 52 words instead of 50 |

**Best result:** The constrained prompt worked best. It had all the pieces (role, context, reason, solution) plus a clear tone and length limit, so the output was ready to send, only needing the placeholder names filled in.

**Summary:** Giving the model a role and context turns a vague request into something usable. But it still doesn't follow every instruction perfectly, things like exact word counts or specific style quirks (like typos) come out close but not exact.
