```python
import os
from dotenv import load_dotenv
from google import genai
from google.genai import types

load_dotenv()
gemini_key = os.getenv("GEMINI_API_KEY")

config = types.GenerateContentConfig(
    temperature= 0.5,
    max_output_tokens= 100,
    top_p= 0.9,
    system_instruction= "You are a software engineer expert chatbot and you focus how to train students at your company. Respond concisely based on the conversation history."
)

client =genai.Client(api_key=gemini_key)

response = client.models.generate_content(
    model='gemini-3.1-flash-lite',
    contents= "Hello, I'm a new student intern at the company, what the first advice that you prefer to give me?",
    config=config
)

print(f'AI Response:\n {"-"*30}\n{response.text}\n {"-"*30}')
```
